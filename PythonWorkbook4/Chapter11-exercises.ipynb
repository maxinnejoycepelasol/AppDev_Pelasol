- What are the constants in the math module?

- What's does math.exp do? What about math.pow?

- How do you calculate logarithms? Logarithms base 10? Natural Logarithms?


```python
import math
help(math)

# Constants in the math module
print(f"Value of pi: {math.pi}")
print(f"Value of e: {math.e}")

# math.exp function
x = 2
print(f"Exponential of {x} is:", math.exp(x))

# math.pow function
x = 2
y = 9
print(f"{x} raised to the power of {y} is:", math.pow(x, y))

# Logarithms
x = 2
number = 202
print(f"Logarithm of {x} with base e is: {math.log(x)}")
print(f"Log base 10 of {number} is: {math.log10(number)}")
print(f"Natural logarithm of e is: {math.log(math.e)}")
```

```python
%%writefile mymodule.py

def calculate_grade(score):
    """
    Calculate/determine the grade based on the score.

    Parameters:
    score (float): The score obtained by the student.

    Returns:
    str: The grade corresponding to the score.
    """
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
```

```python
import mymodule


score1 = 69
score2 = 92

grade1 = mymodule.calculate_grade(score1)
grade2 = mymodule.calculate_grade(score2)

print(f"Score 1: {score1}, Grade: {grade1}")
print(f"Score 2: {score2}, Grade: {grade2}")
```
